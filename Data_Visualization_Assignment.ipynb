# Question
**Transform the Iris dataset into 3 dimensions and plot a 3D chart with the transformed dimensions. Colour each data point with the corresponding classes.**

```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```
# loadingt the data
iris_data = datasets.load_iris()
iris_data
```

```
# transformation of the data into 3 dimensions using PCA
X_reduced = decomposition.PCA(n_components=3).fit_transform(iris_data.data)
```

```
# the reduced data
X_reduced
```

```
# 3D plotting

fig = plt.figure(figsize=(10, 8))
ax = Axes3D(fig)

ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2],
           c=iris_data.target, s=60)

ax.set_title('First Three PCA Directions of Iris Data')
ax.set_xlabel('1st eigenvector')
ax.set_ylabel('2nd eigenvector')
ax.set_zlabel('3rd eigenvector')

plt.show()
```
